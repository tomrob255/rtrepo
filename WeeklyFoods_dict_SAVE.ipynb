```python
# General class 'Food' definition

#Food proteins: g /100g
#Food energy: kcal /100g

class Food:
    counter = 0
    
    def __init__(self, name, energy = 0, proteins = 0):
        self.counter += 1
        Food.counter += 1
        self.name = name
        self.energy = energy
        self.proteins = proteins
    
    def getName(self):
        return self.name
    
    def getProteins(self):
        return self.proteins
    
    def getEnergy(self):
        return self.energy

    @classmethod
    def getFoodsCounter(cls):
        return cls.counter
```

```python
#FoodList with dictionary

class FoodList():
    foodDict = {}
    foodsCounter = 0
    
    def __iter__(self):
        self.iterCounter = 0
        self.keyList = [key for key in self.foodDict.keys()]
        self.foodsCounter = len(self.keyList)
        return self
    
    def __next__(self):
        if self.iterCounter < self.foodsCounter:
            self.iterCounter += 1
            return self.foodDict[self.keyList[self.iterCounter-1]]
        else:
            raise StopIteration    
    
    
    def addFood(self, name, proteins, energy):
        self.foodDict[name] = Food(name, proteins, energy)
        self.foodsCounter += 1
    
    def delAllFoods(self):
        self.foodDict.clear()
                        
    def showFoodsName(self):
        print("Foods names:\n")
        print("Name\t\tProteins\tEnergy")
        for f in self:
            print(f.name+ "\t\t" + str(f.proteins) + "\t\t" + str(f.energy))
            
    def getFoodsList(self):
        names = [ele for ele in self.foodDict.keys()]
        return names
    
    def saveFile(self, path):
        with open(path, mode = 'w') as csv_file:
            fieldnames = ['name', 'proteins', 'energy']
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
            writer.writeheader()
            for food in self:
                writer.writerow({'name': food.name, 'proteins': food.proteins, 'energy': food.energy})
        print("FoodList save to: " + path)
    
    def readFile(self, path):
        with open(path, mode = 'r') as csv_file:
            reader = csv.DictReader(csv_file)
            rowCounter = 0
            for row in reader:
                if rowCounter > 0:
                    self.addFood(row['name'], row['proteins'], row['energy'])
                rowCounter += 1
            self.foodsCounter += rowCounter-1
            print(f"Added {rowCounter-1} elements to FoodList")

            
    
myFoodList = FoodList()
myFoodList.addFood("pollo", 5, 20)
myFoodList.addFood("pasta", 1, 50)
myFoodList.addFood("albume", 100, 10)
myFoodList.addFood("verdura", 1, 1)
myFoodList.addFood("pesce", 50, 10)
myFoodList.addFood("noci", 10, 1)
myFoodList.addFood("latte", 3, 10)
myFoodList.addFood("frutta", 1, 5)
#myFoodList.saveFile('myFoodList.txt')

#myFoodList.readFile('myFoodList.txt')
#myFoodList.delAllFoods()

myFoodList.showFoodsName()

```

```python
#Portion and Meal class definitions

class Portion():
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity

class Meal():
    portionsList = []
    portionsCounter = 0
    
    def __iter__(self):
        self.iterCounter = 0
        return self
    
    def __next__(self):
        if self.iterCounter < self.portionsCounter:
            self.iterCounter += 1
            return self.portionsList[self.iterCounter-1]
        else:
            raise StopIteration
    
    def addPortion(self, name, quantity):
        self.portionsList.append(Portion(name, quantity))
        self.portionsCounter += 1
            
    def showPortions(self):
        print("Meal portions")
        print("Name\t\tQuatity")
        for portion in self:
            print(portion.name + "\t\t" + str(portion.quantity))
    
    def getMealProteins(self, foodList):
        proteins = 0
        foodListNames = foodList.getFoodsList()
        for portion in self:
            if portion.name in foodListNames:
                proteins += foodList.foodDict[portion.name].getProteins()*portion.quantity/100
            else:
                print("Food name: " + portion.name + " not founded!")
        return proteins
    
    def getMealEnergy(self, foodList):
        energy = 0
        foodListNames = foodList.getFoodsList()
        for portion in self:
            if portion.name in foodListNames:
                energy += foodList.foodDict[portion.name].getEnergy()*portion.quantity/100
            else:
                print("Food name: " + portion.name + " not founded!")
        return energy
        
# myMeal = Meal()
# myMeal.addPortion("pollo", 150)
# myMeal.addPortion("pasta", 80)
# myMeal.addPortion("verdura", 250)

# myMeal.showPortions()
# print("myMeal proteins:\t" + str(myMeal.getMealProteins(myFoodList)) + " g")
# print("myMeal energy:\t" + str(myMeal.getMealEnergy(myFoodList)) + " kcal")
```

```python
class Week:
    daysList = []
    daysCounter = 0
    
    def __iter__(self):
        self.iterCounter = 0
        return self

    def __next__(self):
        if self.iterCounter < self.daysCounter:
            self.iterCounter += 1
            return self.daysList[self.iterCounter-1]
        else:
            raise StopIteration
    
    def addDay(self, name):
        self.daysList.append(Day(name))
        self.daysCounter += 1
        
    def showDays(self):
        print("Week:")
        for index in range(self.daysCounter):
            print(str(index) + "\t" + self.daysList[index].name)
    
    def saveFile(self, path):
        with open(path, mode = 'r') as csv_file:
            fieldnames = ['DayNumber', 'DayName', 'MealNumber', 'PortionName', 'PortionQuantity']
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
            writer.writeheader()
            dayNumber = -1
            for day in self:
                dayNumber += 1
                mealNumber = -1
                for meal in day:
                    mealNumber += 1
                    for portion in meal:
                        writer.writerow({'DayNumber': dayNumber, 'DayName' = day.name, 'MealNumber' : mealNumber, 'PortionName': portion.name, 'PortionQuantity': portion.quantity})
                print("Day " + str(dayNumber) +"\t - " + day.name + " completed!")

            
            
    
myWeek = Week()
myWeek.addDay('LUN')
myWeek.showDays()
```
