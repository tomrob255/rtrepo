```python
#Test

class MyClass():
    def __init__(self, value):
        self.value  = value
    def getValue(self):
        return self.value

a = {}
a['primo']     = MyClass('casa')
a['secondo'] = MyClass('albero')
a['terzo']      = MyClass('macchina')

    
for element in a:
    print(a[element].getValue())
    
keys = [el for el in a.keys()]
len(keys)

keys.insert(1, 'ciao')
print(keys)
del(keys[1])
print(keys)
```
